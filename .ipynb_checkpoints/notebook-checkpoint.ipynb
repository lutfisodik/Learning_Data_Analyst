# Exploratory Data Analysis (EDA)

### Analisis Musiman dan Trend Penggunaan Sepeda

- Pengaruh Musim (Season)

Pertama mari kita lihat rata-rata jumlah pengguna sepeda permusim 
dengan kode berikut

```python
average_by_season = day_df.groupby('season')['cnt'].mean()
print(average_by_season)
```

- Trend Tahunan, Bulanan, dan Harian

```python
# Penggambaran penggunaan Sepeda Pertahun
yearly_trend = day_df.groupby('yr')['cnt'].sum()
print(yearly_trend)
```

```python
# Penggambaran Penggunaan sepeda perbulan
monthly_trend = day_df.groupby('mnth')['cnt'].mean()
print(monthly_trend)
```

```python
# Penggambaran Penggunaan sepeda perhari
daily_trend = day_df.groupby('weekday')['cnt'].mean()
print(daily_trend)
```

### Analisis Pengaruh Cuaca Terhadap Penggunaan Sepeda

Oke, sekarang mari kita lihat pengaruh cuaca terhadap penggunaan sepeda dengan 
melihat rata-rata jumlah pengguna sepeda per kondisi cuaca.

```python
average_by_weather = day_df.groupby('weathersit')['cnt'].mean()
print(average_by_weather)
```

- Pengaruh Suhu, Kelembapan, dan kecepatan angin (temp, hum, windspeed)

Ditahap ini kita melihat korelasi antara variabel suhu, kelembapan, kecepatan angin dengan

### Pertanyaan 6

Untuk melihat persentase perbedaan antara jumlah pengguna pada hari libur dan hari kerja

## Analisis RFM (recency, frequency, dan monetary)

Pada tahap ini kita akan melakukan metode analisis RFM untuk melihat kapan terakhir kali pengguna melakukan transaksi, melihat frekuensi dari penggunaan sepeda serta mengindetifikasi nilai transaksi penggunan sepeda.

Mari kita mulai dengan menggabungkan kedua dataframe day.csv dengan hour.csv kolom yang relevan seperti dteday, casual, registered, dan cnt yang menunjukkan jumlah total pengguna. Setelah itu, kita akan menghitung Recency, Frequency, dan Monetary untuk masing-masing pengguna.

```python
# Menggabungkan kedua dataframe berdasarkan kolom yang relevan
merged_df = pd.merge(day_df[['dteday', 'casual', 'registered', 'cnt']], hour_df[['dteday', 'casual', 'registered', 'cnt']], on='dteday')
```

```python
merged_df.columns
```

```python
# Menghitung Recency, Frequency, dan Monetary
recency = merged_df.groupby('dteday')['dteday'].count().reset_index(name='Recency')

# Frequency: Jumlah total penggunaan sepeda (casual + registered)
frequency = merged_df.groupby('dteday')['casual_x'].sum().reset_index()
frequency['Frequency'] = frequency['casual_x']
frequency = frequency[['dteday', 'Frequency']]

monetary = merged_df.groupby('dteday')['cnt_x'].sum().reset_index(name='Monetary')

rfm_analysis = pd.merge(recency, frequency, on='dteday')
rfm_analysis = pd.merge(rfm_analysis, monetary, on='dteday')

# Menampilkan hasil RFM Anaysis
print(rfm_analysis)
```

```python
print(rfm_analysis.describe())
```

### Insight Hasil Penerapan Metode Analisis RFM

Setelah melakukan proses analisis RFM, berdasarkan hasil rangkuman statistika diatas bahwa

- (Recency)
      -Rata-rata waktu terakhir transaksi penggunaan sepeda tercatat sekitar 23-24 hari yang lalu, dengan sebagian besar transaksi memiliki nilai Recency sekitar 24 hari. Terdapat beberapa transaksi yang baru saja dilakukan, terindikasi dari nilai minimum 1.

- (Frequency)
      -Rata-rata frekuensi penggunaan sepeda menunjukkan angka sekitar 20.300 transaksi, dengan sebagian besar data berada pada rentang tersebut. Frekuensi pembelian berkisar dari 2 hingga 81.840, yang mengindikasikan adanya hari-hari dengan frekuensi pembelian sangat rendah hingga sangat tinggi.

- (Monetary)
      -Nilai rata-rata transaksi sekitar 107.785, dengan sebagian besar data berada di sekitar nilai tersebut. Rentang nilai transaksi berkisar dari 22 hingga 209.136. Dengan median nilai transaksi sekitar 108.984, mayoritas transaksi memiliki nilai belanja yang seragam.

 

 Secara keseluruhan, sebagian besar transaksi terjadi dalam rentang waktu 24 hari, dengan mayoritas transaksi berada pada frekuensi dan nilai belanja yang konsisten. Terdapat variasi yang signifikan dalam frekuensi dan nilai belanja yang menandakan variasi pola pembelian pelanggan dari rendah hingga tinggi.

- Visualisasi Recency

```python
plt.figure(figsize=(8, 6))
plt.bar(rfm_analysis['dteday'], rfm_analysis['Recency'], color='skyblue')
plt.xlabel('Date')
plt.ylabel('Recency')
plt.title('Recency Analysis')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- Visualisasi Frequency

```python
plt.figure(figsize=(8, 6))
plt.plot(rfm_analysis['dteday'], rfm_analysis['Frequency'], marker='o', color='green')
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.title('Frequency Analysis')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- Visualisasi Monetary

```python
plt.figure(figsize=(8, 6))
plt.bar(rfm_analysis['dteday'], rfm_analysis['Monetary'], color='orange')
plt.xlabel('Date')
plt.ylabel('Monetary')
plt.title('Monetary Analysis')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
